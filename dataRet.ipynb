```python
#build url
search_term = 'nba'

another_term = 'basketball'

url = f'https://api.twitter.com/2/tweets/search/recent?query={search_term}%20{another_term}&max_results=100'



user_fields = f'&user.fields=created_at,id,location,name,public_metrics,url,username'

url += user_fields

tweet_fields = f'&tweet.fields=attachments,entities,geo,id,lang,public_metrics,source,text'

url += tweet_fields

attachments = f'&expansions=attachments.poll_ids,attachments.media_keys,author_id,geo.place_id,referenced_tweets.id,entities.mentions.username,referenced_tweets.id.author_id'

url += attachments

media_fields = f'&media.fields=duration_ms,height,media_key,preview_image_url,public_metrics,type,url,width'

url += media_fields

country = f'&place.fields=contained_within,country,country_code,full_name,geo,id,name,place_type'

url += country

poll = f'&poll.fields=duration_minutes,end_datetime,id,options,voting_status'

url += poll
```

```python
#request and turn into json_response
response = requests.request("GET", url, headers=headers)
json_response = json.loads(response.text)
new_data = json.dumps(json_response, indent=2, sort_keys=True)
print(new_data)
```

```python
#create dataframe
df = pd.DataFrame(columns=['id', 'lang', 'retweets', 'quotetweets', 'likecount', 'replycount', 'url', 'text'])
df.set_index('id', inplace=True)

for tweet in json_response['data']:
  mediaKey = ""  

  # for loop/store variables
  tweet_id = tweet['id']
  lang = tweet['lang']
  #location = tweet['users']['location']
  retweet_count = tweet['public_metrics']['retweet_count']
  quote_count = tweet['public_metrics']['quote_count']
  like_count = tweet['public_metrics']['like_count']
  reply_count = tweet['public_metrics']['reply_count']
  image_url = ""
  text = tweet['text']

  # Look for picture
  if 'attachments' in tweet:
    if 'media_keys' in tweet['attachments']:
      mediaKey = tweet['attachments']['media_keys'][0]

      #find picture
      if mediaKey != "":
        for media in json_response['includes']['media']:
          if media['media_key'] == mediaKey: 
            if media['type'] == 'photo':      
              image_url = media['url']

              # Add to datafarme
              df.loc[tweet_id] = [lang, retweet_count, quote_count, like_count, reply_count, image_url, text]

df.head()
```
